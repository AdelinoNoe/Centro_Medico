```python
#eliminando os espaços no início e no final das colunas 
df.columns= df.columns.str.strip()
```

```python
#Renomeando as colunas
df.rename(columns={'Valor Exame':'Valor_Exame','Valor Consulta':'Valor_Consulta', 'Valor Exame':'Valor_Exame',  'Valor Tratamento':'Valor_tratamento', 'DÍVIDAS':'Divida'}, inplace= True)
```

```python
#Copiando a tabela original para testes e tratamentos de dados
df_limpesa1=df.copy()
```

```python
df_limpesa1.head()
```

```python
#Visusalização de elementos String com base a coluna Idade 
df_string=df_limpesa1[df['Idade'].astype(str).str.contains(r"[A-Za-z]", regex=True)]
df_string
```

```python
#Tartamento do NaN para os tipos int
cols = ['Valor_Consulta','Valor_Exame','Valor_tratamento','Divida','Idade']
df_limpesa1[cols] = df_limpesa1[cols].fillna(0)
```

```python
df_limpesa1.head(10)
```

```python
df_limpesa1.isnull().sum()
```

```python
#Tratamentos dos NaN para os tipos String
coluna_df_Tratamento=['sexo','Consulta','Tratamento','Descrição','Exames']
df_limpesa1[coluna_df_Tratamento]= df_limpesa1[coluna_df_Tratamento].fillna("Não Informado")
```

```python
#df_teste= df_limpesa1.copy()
#df_teste['Divida']= df_teste['Divida'].astype(float)
```

```python
#Conversão para os respectivos tipos de dados 
#Tratamento de tipo de dado sting-float,data para o tipo DataTime, e idade para int
#df_teste['Divida']= pd.to_numeric(df_teste['Divida'], errors="coerce")
df_limpesa1['Divida']= pd.to_numeric(df_limpesa1['Divida'], errors="coerce")
df_limpesa1['Data']= pd.to_datetime(df_limpesa1['Data'],dayfirst=True,errors="coerce")
df_limpesa1['Idade']= pd.to_numeric(df_limpesa1['Idade'], errors="coerce")
```

```python
df_limpesa1.info()
```

```python
df_limpesa1.info()
```

```python
#Transformado os NaN gerados pela conversão em 0
df_limpesa1['Divida'] = df_limpesa1['Divida'].fillna(0)
df_limpesa1['Idade'] = df_limpesa1['Idade'].fillna(0)
```

```python
df_limpesa1['Idade']=df_limpesa1['Idade'].astype(int)
```

```python
df_limpesa1['sexo']= (
    df_limpesa1['sexo']
    .str.replace('F', 'Femenino')
    .str.replace('M','Masculino')
    )

#df_limpesa1['sexo'].replace({'M':'Masculino', 'F':'Femenino'}, inplace=True)
```

```python
df_limpesa1.head()
```

```python
#eliminação dos espaços no início e final no DataFrame
for col in df_limpesa1.select_dtypes(include=['object']).columns:
    df_limpesa1[col]= df_limpesa1[col].str.strip()
```
